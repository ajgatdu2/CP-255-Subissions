### 3). Creating centroids from barangay geometry

```python
cent = mnl.centroid
cent.plot(markersize = 1, figsize=(10,10))
```

```python
cent.type()
```

```python
#Mapping centroids on top of barangays to see 

f, ax = plt.subplots(1, figsize=(10,10))
mnl.plot(ax=ax, color = 'blue', alpha = 0.5, markersize=1, figsize=(10,10))
cent.plot(ax=ax, color = 'blue', markersize = 0.5, figsize=(10,10))

plt.show()
```

```python
#Plotting barangay centroids on walking network

f, ax = plt.subplots(1,1, figsize=(10,10))
gdf.plot(ax=ax, color = 'red', alpha = 0.3, markersize=1, figsize=(10,10))
cent.plot(ax=ax, color = 'blue', markersize =4, figsize=(10,10), zorder=2)

plt.show()
```
